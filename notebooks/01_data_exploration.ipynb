```python
import pandas as pd

df = pd.read_csv('../data/raw/GiveMeSomeCredit-training.csv')
df.head()
```

```python
df[['RevolvingUtilizationOfUnsecuredLines', 
    'DebtRatio',
    'NumberOfTime30-59DaysPastDueNotWorse',
    'NumberOfTimes90DaysLate',
    'NumberOfTime60-89DaysPastDueNotWorse',
    'NumberOfDependents',
    'age']].describe()
```

```python
# 범위 벗어나는 행 제거
df = df[
    (df['RevolvingUtilizationOfUnsecuredLines'] >= 0) & (df['RevolvingUtilizationOfUnsecuredLines'] <= 1) &
    (df['DebtRatio'] >= 0) & (df['DebtRatio'] <= 1) &
    (df['NumberOfTime30-59DaysPastDueNotWorse'] >= 0) & (df['NumberOfTime30-59DaysPastDueNotWorse'] <= 50) &
    (df['NumberOfTimes90DaysLate'] >= 0) & (df['NumberOfTimes90DaysLate'] <= 50) &
    (df['NumberOfTime60-89DaysPastDueNotWorse'] >= 0) & (df['NumberOfTime60-89DaysPastDueNotWorse'] <= 50) &
    (df['NumberOfDependents'] >= 0) & (df['NumberOfDependents'] <= 10) &
    (df['age'] >= 18) & (df['age'] <= 100)
]

print(f'이상값 제거 후 데이터 수: {df.shape[0]}')
```

```python
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)
df['NumberOfDependents'].fillna(df['NumberOfDependents'].median(), inplace=True)

df.isnull().sum()
```

```python
df = df.fillna({
    'MonthlyIncome': df['MonthlyIncome'].median(),
    'NumberOfDependents': df['NumberOfDependents'].median()
})

df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.to_csv('../data/processed/credit_cleaned.csv', index=False)
print('Saved!')
```
